```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

# **Uploading the Image**

```python
original = cv2.imread("/content/drive/MyDrive/messi.jpg")
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(original, cmap='gray')
```

# **Converting The Image To Gray Scale**

```python
grayScaleImage = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
```

```python
grayScaleImage
```

```python
grayScaleImage.shape
```

```python
resize = cv2.resize(grayScaleImage, (960, 540))
```

```python
plt.imshow(resize, cmap='gray')
```

# **Blurring The Image**

```python
smooth = cv2.medianBlur(grayScaleImage, 5)
resize2 = cv2.resize(smooth, (960, 540))
```

```python
plt.imshow(resize2,cmap='gray')
```

# **Edge Detection**

```python
getEdge = cv2.adaptiveThreshold(smooth, 255, 
  cv2.ADAPTIVE_THRESH_MEAN_C, 
  cv2.THRESH_BINARY, 9, 9)
```

```python
resize3 = cv2.resize(getEdge, (960, 540))
plt.imshow(resize3, cmap='gray')
```

# **Mask Image**

```python
#applying bilateral filter to remove noise 
#and keep edge sharp as required
colorImage = cv2.bilateralFilter(original, 9, 300, 300)
resize4 = cv2.resize(colorImage, (960, 540))
plt.imshow(resize4, cmap='gray')
```

# **Cartoonify**

```python
cartoon = cv2.bitwise_and(colorImage, colorImage, mask=getEdge)
resize5 = cv2.resize(cartoon, (960, 540))
plt.imshow(resize5, cmap='gray')
```
