# Train test split

```python
X_train = train.iloc[:, :-1].values
y_train = train.iloc[:, 132].values

X_test = test.iloc[:, :-1].values
y_test = test.iloc[:, 132].values
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

# **ModelTraining**

```python
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Decision Tree

```python
dt_clf = DecisionTreeClassifier(splitter='best', criterion='entropy', min_samples_leaf=2)


dt_clf.fit(X_train, y_train)
```

```python
pred = dt_clf.predict(X_test)
```

```python
pd.DataFrame({
    'Actual': y_test,
    'Predicted': pred
}).head(10)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, pred))
```

# Random Forest

```python
rf_clf = RandomForestClassifier(criterion='entropy', min_samples_leaf=2)


rf_clf.fit(X_train, y_train)
```

```python
pred_rf = rf_clf.predict(X_test)
```

```python
pd.DataFrame({
    'Actual': y_test,
    'Predicted': pred_rf
}).head(10)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, pred_rf))
```

# Multilayer Perceptron

```python

mlp_clf = MLPClassifier()


mlp_clf.fit(X_train, y_train)
```

```python
pred_mlp = mlp_clf.predict(X_test)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, pred_mlp))
```

```python
pd.DataFrame({
    'Actual': y_test,
    'Predicted': pred_mlp
}).head(10)
```
