```python
from textblob import TextBlob
```

```python
text =  "i willa learna to readgf "

blob = TextBlob(text)
correct_text = blob.correct()
print(correct_text)
```

```python
def correct_spelling(word):
  correct_word = []
  for i in word:
    correct_word.append(TextBlob(i))

  print("Wrong words:", word)
  corrected = []
  for i in correct_word:
    corrected.append(i.correct())
  return corrected
```

```python
words = ["learna","learnedsd"]
out = correct_spelling(words)
print(out)
```
