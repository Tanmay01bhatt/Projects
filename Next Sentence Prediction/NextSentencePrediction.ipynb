```python
import pandas as pd
import random
```

```python
data = pd.read_csv("/content/fake_or_real_news.csv")
data.head()
```

```python
text = list(data.text.values)

joined_text = " ".join(text)# concatenate a list of text elements into a single string with spaces between them
```

```python
with open("joined_text.txt", "w", encoding="utf-8") as f:  #save text or data to a file
   f.write(joined_text)                                    #  opening a file named "joined_text.txt" in write mode with UTF-8 encoding
```

```python
partial_text = joined_text[:100000]
partial_text
```

# Tokenize

```python
import numpy as np
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r"\w+")
tokens = tokenizer.tokenize(partial_text.lower())
```

```python
tokens
```

```python
unique_tokens = np.unique(tokens)
```

```python
unique_token_index = {token: index for index, token in enumerate(unique_tokens)}
```

```python
unique_token_index
```

```python
len(tokens)
```

```python
#Finding prev input words and next words
n_words = 10
input_words = []
next_word = []

for i in range(len(tokens)-n_words):
  input_words.append(tokens[i:i+n_words])
  next_word.append(tokens[i+n_words])


print("Previous Input Words :",input_words[0])
print("Next Word :",next_word[0])
```

```python
#Creating the array filled with number of zeros
X= np.zeros((len(input_words), n_words, len(unique_tokens)), dtype=bool)
y =np.zeros((len(next_word), len(unique_tokens)), dtype=bool)
```

```python
X.shape,y.shape
```

```python
#Properly indexing the previous and next words
for i, words in enumerate(input_words):
    for j, word in enumerate(words):
        X[i, j, unique_token_index[word]] = 1
    y[i, unique_token_index[next_word[i]]] = 1
```

```python
y
```

# Model Buiding

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Activation
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(n_words, len(unique_tokens)), return_sequences=True))
model.add(LSTM(128))
model.add(Dense(len(unique_tokens)))
model.add(Activation("softmax"))
```

```python
optimizer = RMSprop(learning_rate=0.01)
model.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
history = model.fit(X, y, batch_size=128, epochs=50, shuffle=True).history
```

# OUTPUT

```python
def predict_next_word(input_text, n_best):
    input_text = input_text.lower()
    X = np.zeros((1, n_words, len(unique_tokens)))# 1 input sentence
    for i, word in enumerate(input_text.split()):
        X[0, i, unique_token_index[word]] = 1   #check if each word is present in unique words index

    predictions = model.predict(X)[0]
    return np.argpartition(predictions, -n_best)[-n_best:]
```

```python
possible = predict_next_word("I will have to look into this thing because I", 5)
```

```python
for idx in possible:
    print(unique_tokens[idx])
```

```python
def generate_text(input_text, n_words, most_possible_choice=3):
    word_sequence = input_text.split()
    current = 0
    for _ in range(n_words):
        sub_sequence = " ".join(tokenizer.tokenize(" ".join(word_sequence).lower())[current:current+n_words])
        try:
            choice = unique_tokens[random.choice(predict_next_word(sub_sequence,most_possible_choice))]
        except:
            choice = random.choice(unique_tokens)
        word_sequence.append(choice)
        current += 1
    return " ".join(word_sequence)
```

```python
generate_text("I will have to look into this thing because I", 50, 5)
```
